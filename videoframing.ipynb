```python
from imutils import face_utils
from google.colab.patches import cv2_imshow
import numpy as np
import imutils
import dlib
import cv2
import math


detector = dlib.get_frontal_face_detector()

def extract_image_one_fps(video_source_path):

    vidcap = cv2.VideoCapture(video_source_path)
    count = 0
    real_count = 0
    success = True
    while success:
      vidcap.set(cv2.CAP_PROP_POS_MSEC,(count*16))
      success,image = vidcap.read()
      image = cv2.rotate(image, cv2.ROTATE_90_CLOCKWISE)
      
      ## Stop when last frame is identified
      image_last = cv2.imread("frame{}.png".format(count-1))
      if np.array_equal(image,image_last):
          break

      gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
      rects = detector(gray, 1)

      for (i, rect) in enumerate(rects):
        (x, y, w, h) = face_utils.rect_to_bb(rect)
        if(count == 0): 
          width=w
          height = h
        image = image[y:y+height,x:x+width]

      check_height, check_width = image.shape[:2]
      if (check_height==height and check_width==width):
        cv2.imwrite("/gdrive/My Drive/videoimage3/frame%d.png" % count, image)     # save frame as PNG file
        print('{}.sec reading a new frame: {} '.format(count,success))
      count += 1

extract_image_one_fps('/gdrive/My Drive/video3.mp4')
```
